```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
df = pd.read_csv("weightdataset.csv", sep = ';')
```

```
n = 10000
x = df.iloc[:, 1]
y = df.iloc[:, 2]
x = (x-x.mean())/x.std()
XS = x
```

```
X = np.c_[np.ones(n),x]
```

```
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
```

```
Y = theta[0] + theta[1]*XS
```

```
plt.scatter(x,y ,label ='Scatter Plot', color='b',marker = '*')
plt.plot(XS,Y, 'r',label = 'Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Least Square Regression')
plt.legend()
plt.show()
```
